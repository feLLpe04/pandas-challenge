```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path    
import os
print(os.getcwd())
```

```python
csv_path1 = "Resources/schools_complete.csv"
csv_path2 = "Resources/students_complete.csv"
```

```python
schools_df = pd.read_csv(csv_path1)
students_df = pd.read_csv(csv_path2)
```

```python
total_schools = schools_df["School ID"].count()
total_schools
```

```python
total_students = students_df["Student ID"].count()
total_students
```

```python
total_budget = schools_df["budget"].sum()
total_budget
```

```python
average_math_score = students_df["math_score"].mean()
average_math_score
```

```python
average_reading_score = students_df["reading_score"].mean()
average_reading_score
```

```python
students_passing_math = students_df.loc[students_df["math_score"] >= 70, "math_score"].count()/total_students*100
students_passing_math
```

```python
passing_reading_count = students_df[students_df['reading_score'] >= 70].count()['student_name']
passing_reading_percentage = (passing_reading_count / students_df['student_name'].count()) * 100
passing_reading_percentage
```

```python
overall_passing_count = students_df[(students_df['math_score'] >= 70) & (students_df['reading_score'] >= 70)].count()['student_name']
overall_passing_percentage = (overall_passing_count / students_df['student_name'].count()) * 100
overall_passing_percentage
```

```python
# Merge the students and schools data
school_data_complete = pd.merge(students_df, schools_df, how='left', on='school_name')
```

```python
# Group by school to get the type of each school
school_types = schools_df.set_index('school_name')['type']
```

```python
# Calculate total students per school
total_students_per_school = school_data_complete['school_name'].value_counts()
```

```python
# Calculate total school budget
total_budget_per_school = schools_df.set_index('school_name')['budget']
```

```python
# Calculate per student budget
per_student_budget = total_budget_per_school / total_students_per_school
```

```python
# Calculate average math scores
average_math_scores = school_data_complete.groupby('school_name')['math_score'].mean()

# Calculate average reading scores
average_reading_scores = school_data_complete.groupby('school_name')['reading_score'].mean()
```

```python
# Calculate the number of students passing math per school
passing_math = school_data_complete[school_data_complete['math_score'] >= 70]
passing_math_percentage = passing_math.groupby('school_name')['student_name'].count() / total_students_per_school * 100
```

```python
# Calculate the number of students passing reading per school
passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70]
passing_reading_percentage = passing_reading.groupby('school_name')['student_name'].count() / total_students_per_school * 100
```

```python
# Calculate the number of students passing both math and reading per school
overall_passing = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]
overall_passing_percentage = overall_passing.groupby('school_name')['student_name'].count() / total_students_per_school * 100
```

```python
# Create a DataFrame to hold all the calculated metrics
school_summary_df = pd.DataFrame({
    'School Type': school_types,
    'Total Students': total_students_per_school,
    'Total School Budget': total_budget_per_school,
    'Per Student Budget': per_student_budget,
    'Average Math Score': average_math_scores,
    'Average Reading Score': average_reading_scores,
    '% Passing Math': passing_math_percentage,
    '% Passing Reading': passing_reading_percentage,
    '% Overall Passing': overall_passing_percentage
})

# Display the DataFrame
school_summary_df
```

```python
# Hihgest Performing Schools (first 5)
top_schools = school_summary_df.sort_values('% Overall Passing', ascending=False)
top_schools.head(5)
```

```python
#Lowest performing all schools (first 5)
bottom_schools = school_summary_df.sort_values('% Overall Passing', ascending=True)
bottom_schools.head(5)
```

```python
# Math Scores by Grade Dataframe List Score students 9th, 10th, 11th, 12th
math_scores_by_grade = students_df.groupby(['school_name', 'grade']).mean(numeric_only=True)['math_score']
math_scores_by_grade = math_scores_by_grade.unstack()
math_scores_by_grade_df = pd.DataFrame(math_scores_by_grade)
```

```python
# Reading Scores by Grade Dataframe List Score students 9th, 10th, 11th, 12th
reading_scores_by_grade = students_df.groupby(['school_name', 'grade']).mean(numeric_only=True)['reading_score']
reading_scores_by_grade = reading_scores_by_grade.unstack()
reading_scores_by_grade_df = pd.DataFrame(reading_scores_by_grade)
```

```python
# Spending group by due categories
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ['<$585', '$585-630', '$630-645', '$645-680']

school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], bins=spending_bins, labels=spending_labels)
```

```python
# Total Budget per School
total_budget_per_school = schools_df.set_index('school_name')['budget']
```

```python
# Budget per Student
per_student_budget = total_budget_per_school / total_students_per_school
```

```python
# Average Math Score per School
average_math_scores = students_df.groupby('school_name')['math_score'].mean()
```

```python
# Average Reading Score per School
average_reading_scores = students_df.groupby('school_name')['reading_score'].mean()
```

```python
# % Passing Math per School
passing_math = students_df[students_df['math_score'] >= 70]
passing_math_percentage = passing_math.groupby('school_name')['Student ID'].count() / total_students_per_school * 100
```

```python
# % Passing Reading per School
passing_reading = students_df[students_df['reading_score'] >= 70]
passing_reading_percentage = passing_reading.groupby('school_name')['Student ID'].count() / total_students_per_school * 100
```

```python
#Percentage Overall Passing on Math and Reading
overall_passing_percentage = (passing_math_percentage + passing_reading_percentage) / 2
```

```python
# Create a DataFrame to hold all the metrics
spending_summary = pd.DataFrame({
    "Total Students": total_students_per_school,
    "Total School Budget": total_budget_per_school,
    "Per Student Budget": per_student_budget,
    "Average Math Score": average_math_scores,
    "Average Reading Score": average_reading_scores,
    "% Passing Math": passing_math_percentage,
    "% Passing Reading": passing_reading_percentage,
    "% Overall Passing": overall_passing_percentage
})
spending_summary
```

```python
#Define the bins and labels for the school size
size_bins = [0, 1000, 2000, 5000]
size_labels = ['Small (<1000)', 'Medium (1000-2000)', 'Large (2000-5000)']
school_summary_df["School Size"] = pd.cut(school_summary_df["Total Students"], bins=size_bins, labels=size_labels)
```

```python
#Calculate the average scores and passing rates by school size
size_math_scores = school_summary_df.groupby('School Size').mean()['Average Math Score']
size_reading_scores = school_summary_df.groupby('School Size').mean()['Average Reading Score']
size_passing_math = school_summary_df.groupby('School Size').mean()['% Passing Math']
size_passing_reading = school_summary_df.groupby('School Size').mean()['% Passing Reading']
size_overall_passing = school_summary_df.groupby('School Size').mean()['% Overall Passing']
```

```python
#Size Summary DataFrame
size_summary_df = pd.DataFrame({
    "Average Math Score": size_math_scores,
    "Average Reading Score": size_reading_scores,
    "% Passing Math": size_passing_math,
    "% Passing Reading": size_passing_reading,
    "% Overall Passing": size_overall_passing
})
size_summary_df
```
